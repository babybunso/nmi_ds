```python
%%time
import sys
import numpy as np

# ML - Optimization (storage, power)
# What is Numpy?
# Why is Numpy fast?


## List 
X = [x for x in range(1000000)]
print(f'Sum: {sum(X)}')

s_X = sys.getsizeof(X)
print(f'Memory X: {s_X/1024/1024} MB\n')

## Generator
X_alt = (x for x in range(1000000))
print(f'Sum: {sum(X_alt)}')

s_X_alt = sys.getsizeof(X_alt)
print(f'Memory X_alt: {s_X_alt} bytes\n')


## Numpy Array | list-like object
X_np_arr = [x for x in range(1000000)]
X_np_arr = np.array(X_np_arr) # creating numpy array
print(f'Sum: {sum(X_np_arr)}')

s_X_arr = sys.getsizeof(X_np_arr)
print(f'Memory X_np_arr: {s_X_arr/1024/1024} MB\n\n\n')
```

```python
%%time

# Aggregations
A = np.array([1, 2, 3, 4, 5])
print(f'data: {A}')
print(f'Sum: {np.sum(A)}')
print(f'Median: {np.median(A)}')
print(f'Mean: {np.mean(A)}')
print(f'Min: {np.min(A)}')
print(f'Max: {np.max(A)}')
print(f'Size: {np.size(A)}')
print(f'Len: {len(A)}\n\n')
```

```python
%%time

# argmin, argmax
B = np.array([x+1 for x in range(1000)])
# print(f'data (before): {B}')
np.random.shuffle(B)
# print(f'data (shuffled): {B}')

min_idx = np.argmin(B)
max_idx = np.argmax(B)

print(f'ArgMin: {min_idx}')
print(f'Data Min: {B[min_idx]}')

print(f'ArgMAx: {max_idx}')
print(f'Data MAx: {B[max_idx]}\n\n')


```

# Terminologies: Linear Algebra for ML

<img src="https://hadrienj.github.io/assets/images/2.1/scalar-vector-matrix-tensor.png" />
<br>

- A `scalar` is a single number
- A `vector` is an array of numbers
- A `matrix` is a 2-D array
- A `tensor` is a n-dimensional array with $n$>2
<br><br>
$A = m x n$
- where m is # of rows, and n is the # of columns
- $R x C$ (shape)
